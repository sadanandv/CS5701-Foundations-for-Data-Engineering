Plot the following functions using Python:
1. Plot the function f(x) = x
2 + 2x + 5 in the range −5 ≤ x ≤ 5.
2. Plot the function f(x) = −x

3. Plot the function g(x) = sin(x) in the range −2π ≤ x ≤ 2π.

##Define Functions

```python
def f1(x):
    return x**2 + 2*x + 5

def f2(x):
    return -x**2 + 2*x + 5

def g(x):
    return np.sin(x)

def h1(x):
    return np.exp(-x)

def h2(x):
    return np.exp(-np.abs(x))

def h3(x):
    return np.exp(-x**2)
```


## Define the x ranges for each plot


```python
x_range_1 = np.linspace(-5, 5, 400)
x_range_2 = np.linspace(-2*np.pi, 2*np.pi, 400)
x_range_3 = np.linspace(-5, 5, 400)
```

## Create individual plots


```python
plt.figure(figsize=(12, 10))

plt.subplot(3, 2, 1)
plt.plot(x_range_1, f1(x_range_1))
plt.title("f(x) = x^2 + 2x + 5")

plt.subplot(3, 2, 2)
plt.plot(x_range_1, f2(x_range_1))
plt.title("f(x) = -x^2 + 2x + 5")

plt.subplot(3, 2, 3)
plt.plot(x_range_2, g(x_range_2))
plt.title("g(x) = sin(x)")

plt.subplot(3, 2, 4)
plt.plot(x_range_3, h1(x_range_3))
plt.title("h1(x) = e^-x")

plt.subplot(3, 2, 5)
plt.plot(x_range_3, h2(x_range_3))
plt.title("h2(x) = e^-|x|")

plt.subplot(3, 2, 6)
plt.plot(x_range_3, h3(x_range_3))
plt.title("h3(x) = e^-x^2")

plt.tight_layout()
```

## Create a combined plot with legend


```python
plt.figure()
plt.plot(x_range_1, f1(x_range_1), label='$f(x) = x^2 + 2x + 5$')
plt.plot(x_range_1, f2(x_range_1), label='$f(x) = -x^2 + 2x + 5$')
plt.plot(x_range_2, g(x_range_2), label='$g(x) = sin(x)$')
plt.plot(x_range_3, h1(x_range_3), label='$h1(x) = e^-x$')
plt.plot(x_range_3, h2(x_range_3), label='$h2(x) = e^-|x|$')
plt.plot(x_range_3, h3(x_range_3), label='$h3(x) = e^-x^2$')
plt.legend()
plt.title("Basic Functions")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

#Problem Set 2: Activation Functions
Plot the commonly used activation functions and their derivatives using Python:
1. Plot the Sigmoid Activation Function and its derivative.
2. Plot the Tanh Activation Function and its derivative.
3. Plot the ReLU Activation Function and its derivative.
4. Plot the Leaky ReLU Activation Function and its derivative.
5. Plot the Exponential Linear Unit Activation Function and its derivative.

## Activation functions and their derivatives

```python
# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def elu_derivative(x, alpha=1.0):
    return np.where(x > 0, 1, alpha * np.exp(x))
```

```python
activation_functions = [
    ("Sigmoid", lambda x: 1 / (1 + np.exp(-x)), lambda x: sigmoid(x) * (1 - sigmoid(x))),
    ("Tanh", np.tanh, lambda x: 1 - np.tanh(x)**2),
    ("ReLU", lambda x: np.maximum(0, x), lambda x: np.where(x > 0, 1, 0)),
    ("Leaky ReLU", lambda x: np.where(x > 0, x, 0.01 * x), lambda x: np.where(x > 0, 1, 0.01)),
    ("ELU", lambda x: np.where(x > 0, x, np.exp(x) - 1), lambda x: np.where(x > 0, 1, np.exp(x))),
]
```

## Define the x range


```python
x_range = np.linspace(-5, 5, 400)
```

## Create individual plots

```python
plt.figure(figsize=(15, 10))
for i, (name, func, derivative) in enumerate(activation_functions):
    plt.subplot(2, 5, i + 1)
    plt.plot(x_range, func(x_range), label=name)
    plt.title(name)
    plt.legend()

    plt.subplot(2, 5, i + 6)
    plt.plot(x_range, derivative(x_range), label='Derivative')
    plt.title(name + ' Derivative')
    plt.legend()

plt.tight_layout()
```


## Create a combined plot with legend


```python
plt.figure(figsize=(12, 6))
for name, func, _ in activation_functions:
    plt.plot(x_range, func(x_range), label=name)

plt.title("Activation Functions")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")

plt.tight_layout()
plt.show()
```

#Problem Set 3: Parametric Plot
Plot parametric equations using Python:
1. Plot the parametric equations x(t) = cos(t), y(t) = sin(t) in the range 0 ≤ t ≤ 2π.
2. Plot the parametric equations x(t) = e
−0.1t
cos(t), y(t) = e
−0.1t
sin(t) in the range 0 ≤ t ≤ 20.

##Parametric equations

```python
def parametric_equation_1(t):
    x = np.cos(t)
    y = np.sin(t)
    return x, y

def parametric_equation_2(t):
    x = np.exp(-0.1 * t) * np.cos(t)
    y = np.exp(-0.1 * t) * np.sin(t)
    return x, y
```


## Define the parameter ranges


```python
t_range_1 = np.linspace(0, 2*np.pi, 400)
t_range_2 = np.linspace(0, 20, 400)
```

## Calculate the parametric values


```python
x_values_1, y_values_1 = parametric_equation_1(t_range_1)
x_values_2, y_values_2 = parametric_equation_2(t_range_2)
```

## Create plots

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x_values_1, y_values_1)
plt.title("Parametric Equations: x(t) = cos(t), y(t) = sin(t)")
plt.xlabel("x")
plt.ylabel("y")

plt.subplot(1, 2, 2)
plt.plot(x_values_2, y_values_2)
plt.title("Parametric Equations: x(t) = e^(-0.1t) * cos(t), y(t) = e^(-0.1t) * sin(t)")
plt.xlabel("x")
plt.ylabel("y")

plt.tight_layout()
plt.show()
```

## Piecewise functions

```python
def f(x):
    return np.where(x < 0, x**2, 2*x)

def g(x):
    return np.piecewise(x, [x < -1, (x >= -1) & (x < 2), x >= 2], [lambda x: x**3, lambda x: -x**2, lambda x: x])
```

## Define the x range


```python
x_range = np.linspace(-3, 3, 400)
```

## Calculate the function values

```python
f_values = f(x_range)
g_values = g(x_range)
```

## Create plots

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x_range, f_values)
plt.title("Piecewise Function f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.subplot(1, 2, 2)
plt.plot(x_range, g_values)
plt.title("Piecewise Function g(x)")
plt.xlabel("x")
plt.ylabel("g(x)")

plt.tight_layout()
plt.show()
```
